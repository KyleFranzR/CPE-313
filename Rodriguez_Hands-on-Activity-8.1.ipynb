## Procedures
Load the necessary libraries

```python
import nltk
import pandas as pd
```

```python
data = pd.read_csv('songdata.csv')
data.head()
```

Find the lyrics by the artist The Beatles

```python
beatles = data[data.artist=='The Beatles']
beatles.head()
```

Check the number of Beatles songs in the data set


```python
data.artist.value_counts().loc[['The Beatles']]
```

Take a look at the first song's lyrics

```python
beatles.iloc[0,3]
```

Remove characters from the lyrics

```python
import re

newline = lambda x: re.sub('\n', ' ', x) # remove \n
alphanumeric = lambda x: re.sub('\w*\d\w*', ' ', x) # remove alphanumeric words

corpus = beatles.text.map(newline).map(alphanumeric)
corpus.head()
```

Create a document term matrix using Count Vectorizer with the stop words turned on to English

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words="english")
X = cv.fit_transform(corpus).toarray()

dt = pd.DataFrame(X, columns=cv.get_feature_names_out()).set_index(beatles.song)
dt.head()
```

Set display options so that the lyrics aren't cut off

```python
pd.set_option('display.max_colwidth', None)
```

```python
beatles[beatles.song=='Imagine'].text
```

Imagine lyrics in Count Vectorizer form


```python
imagine = list(dt.loc['Imagine'])
imagine[:20]
```

Define the cosine similarity calculation

```python
from numpy import dot
from numpy.linalg import norm

cosine = lambda v1, v2: dot(v1, v2) / (norm(v1) * norm(v2))
```

Calculate all similarities and sort by the most similar

```python
similarity = [cosine(imagine, song) for song in X]
sorted(list(zip(similarity, beatles.song)), reverse=True)[1:6]
```

Create the TF-IDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer

cv_tfidf = TfidfVectorizer(stop_words="english")
X_tfidf = cv_tfidf.fit_transform(corpus).toarray()

dt_tfidf = pd.DataFrame(X_tfidf, columns=cv_tfidf.get_feature_names_out()).set_index(beatles.song)
dt_tfidf.head()
```

Calculate all similarities and sort by the most similar using the TF-IDF data

```python
similarity_tfidf = [cosine(imagine, song) for song in X_tfidf]
sorted(list(zip(similarity_tfidf, beatles.song)), reverse=True)[1:6]
```

```python
#type your code here
from numpy import dot
from numpy.linalg import norm
import pandas as pd
import re

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.tag import pos_tag
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

#### Importing dataset

I obtained my dataset here:

> https://www.kaggle.com/datasets/jrobischon/wikipedia-movie-plots

```python
# Use a dataset that consists of text document
# Had to limit dataset to only 10k because session will crash

movies = pd.read_csv('/content/wiki_movie_plots_deduped.csv')[['Title', 'Plot']].head(10000).drop_duplicates(subset=['Title'])
movies.head()
```

```python
movies.info()
```

#### Tokenization for the discussion

```python
# tokenize words

movies['Plot_tknz'] = movies.Plot.map(word_tokenize)
movies.head()
```

```python
# Tokenize (words with remove special characters and numbers)

remove_special = lambda x: re.sub(r'[!?]' ,' ', x)
remove_number = lambda x: re.sub(r'[/d!?]' ,' ', x)

movies['Plot_tknz'] = movies.Plot.map(remove_special).map(remove_number).map(word_tokenize)
movies.head()
```

```python
# Tokenize (words with remove punctuation and all lowercase)

remove_punc = lambda x: re.sub(r'[!?]' ,' ', x.lower())
remove_number = lambda x: re.sub(r'[/d!?]' ,' ', x.lower())

movies['Plot_tknz'] = movies.Plot.map(remove_special).map(remove_number).map(word_tokenize)
movies.head()
```

```python
# Tokenize (words with remove punctuation, all lowercase and remove stop words)

stop_words = lambda x: [y for y in x if y not in stopwords.words('english')]

movies['Plot_tknz'] = movies.Plot_tknz.apply(stop_words)
movies.head()
```

```python
# Tokenize (words with remove punctuation, all lowercase and remove stop words, with speech_tagging)

speech_tag = lambda x: nltk.pos_tag(x)

movies['Plot_tknz_tag'] = movies.Plot_tknz.apply(speech_tag)
movies.head()
```

#### Remove special characters

```python
# Now, I will apply of the process that was made

newline = lambda x: re.sub('\n', ' ', x)
remove_punc = lambda x: re.sub(r'[!?]' ,' ', x.lower())
remove_number = lambda x: re.sub(r'[/d!?]' ,' ', x.lower())

testing = movies.Plot.map(newline).map(remove_punc).map(remove_number)
testing.head()
```

#### Compare the similar document using Count Vector

```python
# Count vectorizer

cv = CountVectorizer(stop_words="english")
X = cv.fit_transform(testing).toarray()

plot_cv = pd.DataFrame(X, columns=cv.get_feature_names_out()).set_index(movies.Title)
```

```python
# Let's find out what movies have "similar" plot to Huckleberry Finn
# I use cosine to calculate the distance

cosine = lambda v1, v2: dot(v1, v2) / (norm(v1) * norm(v2))

movie_hf = list(plot_cv.loc['Huckleberry Finn'])
similarity = [cosine(movie_hf, plot) for plot in X]
```

```python
# 5 Movies with most similar plots

sorted(list(zip(similarity, movies.Title)), reverse=True)[1:6]
```

#### Compare the similar document using TF-IDF Vectorizer

```python
# TF-IDF Vectorizer

tf = TfidfVectorizer(stop_words="english")
y = tf.fit_transform(testing).toarray()

plot_tf = pd.DataFrame(y, columns=tf.get_feature_names_out()).set_index(movies.Title)
```

```python
# Let's find out what movies have "similar" plot to Huckleberry Finn

movie_hf = list(plot_tf.loc['Huckleberry Finn'])

similarity = [cosine(movie_hf, plot) for plot in X]
```

```python
# 5 Movies with most similar plots

sorted(list(zip(similarity, movies.Title)), reverse=True)[1:6]
```

> ***As seen in the results, both have the same top 1 and top 3 (not really a different movie), but the rest don't match.***
